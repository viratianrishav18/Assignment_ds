```python
# 30. Generate dataset with 100 samples for consistency
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

X, y = make_moons(n_samples=100, noise=0.1)  # Ensure both X and y have 100 samples
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()

# 31. Scatter plot of X
plt.scatter(X[:, 0], X[:, 1])
plt.show()

# 32. K-Means clustering
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, n_init=10)
kmeans.fit(X)
labels = kmeans.labels_
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.show()

# 33. PCA for dimensionality reduction
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
plt.scatter(X_pca[:, 0], X_pca[:, 1])
plt.show()

# 34. Standardize features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 35. Logistic Regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X, y)

# 36. Accuracy score
from sklearn.metrics import accuracy_score
y_pred = model.predict(X)
print(accuracy_score(y, y_pred))

# 37. Correlation heatmap
import seaborn as sns
import numpy as np
sns.heatmap(np.corrcoef(X.T), annot=True)
plt.show()

# 38. Train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# 39. Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

# 40. Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)

# 41. Support Vector Machine
from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train, y_train)

# 42. K-Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

# 43. Classification Report
from sklearn.metrics import classification_report
print(classification_report(y_test, rf.predict(X_test)))

# 44. Confusion Matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, rf.predict(X_test)))

# 45. ROC-AUC Score
from sklearn.metrics import roc_auc_score
print(roc_auc_score(y_test, rf.predict_proba(X_test)[:, 1]))

# 46. Pipeline with Scaling and Random Forest
from sklearn.pipeline import Pipeline
pipe = Pipeline([('scaler', StandardScaler()), ('classifier', RandomForestClassifier())])
pipe.fit(X_train, y_train)

# 47. Save model using Pickle
import pickle
with open('model.pkl', 'wb') as f:
    pickle.dump(rf, f)

# 48. Load and use saved model
with open('model.pkl', 'rb') as f:
    loaded_model = pickle.load(f)
print(loaded_model.predict(X_test))
```
