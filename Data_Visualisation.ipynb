**Q1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.**


```python

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

# Define x and y values
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a new plot with title and axis labels
p = figure(title="Sine Wave", x_axis_label="X", y_axis_label="sin(X)")

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2)

# Show the plot
show(p)
```

**Q2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.**

```python

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
from bokeh.models import ColumnDataSource

output_notebook()

# Generate random data for x and y
x = np.random.rand(100) * 10
y = np.random.rand(100) * 10

# Generate random sizes and colors
sizes = np.random.rand(100) * 20 + 10  # Sizes between 10 and 30
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], 100)

# Create a data source for the plot
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a new scatter plot
p = figure(title="Random Scatter Plot", x_axis_label="X", y_axis_label="Y")
p.scatter('x', 'y', size='sizes', color='colors', alpha=0.6, source=source)
show(p)
```

**Q3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.**

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource

output_notebook()

# Define the data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a data source
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a new figure for the bar chart
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label="Fruits", y_axis_label="Counts")

# Add bars to the figure
p.vbar(x='fruits', top='counts', width=0.5, source=source, legend_field="fruits",
       fill_color=factor_cmap('fruits', palette=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'], factors=fruits))

# Show the plot
show(p)
```

**Q4. Create a Bokeh histogram to visualize the distribution of the given data.**

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

# Generate the data and histogram
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a new figure for the histogram
p = figure(title="Histogram", x_axis_label="Value", y_axis_label="Frequency")

# Add a quad glyph for the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.7)

# Show the plot
show(p)
```

**Q5. Create a Bokeh heatmap using the provided dataset.**

data_heatmap = np.random.rand(10, 10)

x = np.linspace(0, 1, 10)

y = np.linspace(0, 1, 10)

xx, yy = np.meshgrid(x, y)

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource
import numpy as np
from bokeh.palettes import Viridis256

output_notebook()

# Generate the data for the heatmap
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the data for use in the ColumnDataSource
source = ColumnDataSource(data=dict(
    x=xx.flatten(),
    y=yy.flatten(),
    value=data_heatmap.flatten()
))

# Create a new figure for the heatmap
p = figure(title="Heatmap", x_axis_label="X", y_axis_label="Y",
           x_range=(0, 1), y_range=(0, 1), tools="")

# Define the color mapper
color_mapper = LinearColorMapper(palette=Viridis256, low=data_heatmap.min(), high=data_heatmap.max())

# Add rectangles for each cell in the heatmap
p.rect(x="x", y="y", width=1/10, height=1/10, source=source,
       fill_color={'field': 'value', 'transform': color_mapper}, line_color=None)

# Add a color bar
color_bar = ColorBar(color_mapper=color_mapper, width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
